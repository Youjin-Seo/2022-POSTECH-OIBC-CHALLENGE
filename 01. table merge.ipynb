## merge

```python
# g

g_grop = []
for i in range(g.shape[0]):
  g_grop.append(str(g.iloc[i]['id']) + str(' / ') + str(g.iloc[i]['date']) + str(' / ') + str(g.iloc[i]['hour']))
```

```python
str(g.iloc[0]['id']) + str(' / ') + str(g.iloc[0]['date']) + str(' / ') + str(g.iloc[0]['hour'])
```

```python
g['group'] = g_grop
```

```python
g
```

```python
g_new = g[['id','date','hour','group','amount']]
g_new
```

```python
# w

w_grop = []
for i in range(w.shape[0]):
  w_grop.append(str(w.iloc[i]['id']-1) + str(' / ') + str(w.iloc[i]['date']) + str(' / ') + str(w.iloc[i]['hour']))
```

```python
w['group'] = w_grop
```

```python
w
```

```python
w_new = w[['id','date','hour','group','temperature','humidity','dew_point','wind_dir',
           'wind_spd','uv_idx','visibility','cloudiness','ceiling','pressure','precip_1h']]
w_new
```

```python
Counter(list(w['group'])).most_common(10)
```
